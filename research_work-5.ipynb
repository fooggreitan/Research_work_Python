```python
import pandas as pd
import seaborn as saw
import matplotlib.pyplot as plt
import random
df = pd.read_csv('D:/california_housing_test.csv')
print(df)
```

```python
saw.scatterplot(data=df, x="population", y="households")
```

```python
saw.relplot(x="longitude", y="median_house_value", kind="line", data=df)
```

```python
saw.histplot(data=df[df['housing_median_age'] > 40], x="median_house_value", hue="housing_median_age")
```
